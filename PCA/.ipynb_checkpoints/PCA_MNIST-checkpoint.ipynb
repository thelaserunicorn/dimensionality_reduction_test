```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_table = pd.read_csv('mnist_train.csv')
data_table.head()
```

```python
data_table.shape
```

```python
label = data_table['label']
data = data_table.drop("label",axis=1)
label.head()
data.head()
```

```python
print(data.shape)
print(label.shape)
```

```python
short_label = label.head(15000)
short_data = data.head(15000)
print('shape of=', short_data.shape)
print('shape of=', short_label.shape)
```

```python
#standardizing the data (column standardization)

from sklearn.preprocessing import StandardScaler
standardized_data = StandardScaler().fit_transform(data)
print(standardized_data.shape)

sample_data = standardized_data
print(sample_data.shape)


```

```python
from sklearn import decomposition
pca = decomposition.PCA()
```

```python
pca.n_components = 2
pca_data = pca.fit_transform(sample_data)
print('reduced dim' , pca_data.shape)
```

```python
import pandas as pd
# attaching the label for each 2-d data point 
pca_data = np.vstack((pca_data.T, label)).T

# creating a new data frame which help us in ploting the result data
pca_df = pd.DataFrame(data=pca_data, columns=("1st_principal", "2nd_principal", "label"))
sns.FacetGrid(pca_df, hue="label", height=6).map(plt.scatter, '1st_principal', '2nd_principal').add_legend()
plt.show()
```

Observation:
data points are overlapping to each other so we are not able to distinguish between the class labels
We can conclude here that visualization using PCA is not suitable for the MNIST dataset.
